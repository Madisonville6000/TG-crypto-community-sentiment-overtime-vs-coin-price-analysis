```python
def clean_text(x):

    # If text is a list
    if isinstance(x, list):
        parts = []

        for item in x:
            if isinstance(item, str):
                parts.append(item)

            elif isinstance(item, dict):
                if "text" in item:
                    parts.append(str(item["text"]))

        return " ".join(parts)

    # If text is dictionary
    elif isinstance(x, dict):
        return str(x)

    # If text is normal string
    elif isinstance(x, str):
        return x

    return ""
```
